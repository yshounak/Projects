```python
import pickle

def savedata(data):
    try:
        filename = "words.txt"
        fileobject = open(filename,'wb')
        pickle.dump(data,fileobject)
        fileobject.close()
        print("SUCCESS: writing to dictionary words.txt")
    except Exception as e:
        print(f"ERROR: {e}")
    
def unpack():
    try:
        filename = "words.txt"
        fileobject = open(filename,'rb')
        data =pickle.load(fileobject)
        fileobject.close()
        print("SUCCESS: reading to dictionary words.txt")
        return data
    except Exception as e:
        print(f"ERROR: {e}")
        

def addnewword():
    print("_"*30)
    print("NEW WORD ADDER")
    print("_"*30)
    word = input("Enter the new word: ")
    meaning = input(f"Enter the meaning of {word}: ")
    dictionary = unpack()
    dictionary[word] = meaning
    savedata(dictionary)
    main_menu()
    
def findmeaning():
    print("_"*30)
    print("FIND MEANING")
    print("_"*30)
    word1 = input("Enter the word to find it's meaning: ")
    dictionary = unpack()
    if word1 in dictionary:
        print(f"Found Entry: [{word1} : {dictionary[word1]}]")
    else:
        print("ERROR: The word doesn't exist!")
    main_menu()

def updateword():
    print("_"*30)
    print("UPDATE WORD MEANING")
    print("_"*30)
    word2 = input("Enter the word to update: ")
    dictionary = unpack()
    if word2 in dictionary:
        updated_meaning = input("Enter a New Meaning to edit [{} : {}] to:".format(word2,dictionary[word2]))
        dictionary[word2] = updated_meaning
        savedata(dictionary)
        print(f"SUCCESS: Updated to [{word2} : {dictionary[word2]}]")
    else:
         print("ERROR: The word does not exist")
    main_menu()
            
def main_menu():
    print("_"*30)
    print("ENGLISH DICTIONARY APP")
    print("_"*30)
    print("1. Add a new word")
    print("2. Find the meaning")
    print("3. Update a word")
    print("4. Exit")
    print("_"*30)
    n = int(input("Enter the number to perform that action: "))
    if n==1:
        addnewword()
    elif n==2:
        findmeaning()
    elif n==3:
        updateword()
    elif n==4:
        exit()
    else:
        print("ERROR: Please Enter a digit according to Menu.")
        main_menu()
        
main_menu()
```
