## Cleaning and preparing the data for model training


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Problem Statement 

A retail company "ABC Private Limited" wants to understand the customer purchase behaviour (specifically, purchase amount) against various product or different categories. They have shared purchase summary of various customers for selected high volume products from last month. The data set also contains cutomer demographics (age, gender, marital status, city_type, stay_in_current_city), products details (product_id and product category) and total purchase amount from last month.

Now, they want to build a model to predict the purchase amount of customer against various pproducts which will help them to create personalized offer for cutomers against different products

```python
# importing the dataset

df_train=pd.read_csv("train.csv")
df_train.head()
```

```python
## import the test data

df_test = pd.read_csv("test.csv")
df_test.head()
```

```python
## Merge both train and test data

df = pd.concat([df_train, df_test], ignore_index=True)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# dropping unuseable columns
df.drop(["User_ID"], axis=1, inplace=True) #axis=1 is for column and inplace=True for updating automatically in df
```

```python
df.head()
```

```python
#Convert Categorical value into numerical

#pd.get_dummies(df["Gender"]) this doesn't allow direct changes to dataframe hence we wont use it
df['Gender']=df['Gender'].map({"F":0,"M":1})
#or df["Gender"]= pd.get_dummies(df["Gender"])
df.head()
```

```python
df.Age.unique()
```

```python
#df.Age = pd.get_dummies(df.Age, drop_first=True) 
# Here domain knowledge will come to help,like 0-17 years will be less, 26-35 will be more and so on
# that is why we'll process it as ordinal values to get ranking 

df.Age = df.Age.map({"0-17":1,"18-25":2,"26-35":3,"36-45":4,"46-50":5,"51-55":6,"55+":7})

#label encoding could also be possible here

#from sklearn import preprocessing
#label_encoder object knows how to understand word labels.
#label_encoder = preprocessing.LabelEncoder()
# Encode labels in column 'species'
#df.Age = label_encoder.fit_transform(df.Age)

df.head()
```

```python
df_city = pd.get_dummies(df["City_Category"], drop_first=True)
df_city = df_city.astype(int)
```

```python
df = pd.concat([df,df_city], axis=1)
df.head()
```

```python
df.drop("City_Category", axis = 1, inplace = True)
```

```python
df.head()
```

```python
# Checking the missing values

df.isnull().sum()
```

```python
# Focus on replacing prod cat 2 and 3 not purchase as it is part of test data

df["Product_Category_2"].unique() #discrete feature
```

```python
df["Product_Category_2"].value_counts()
```

```python
#In discrete feature how do you change nan value? - We'll use mode

df["Product_Category_2"] = df["Product_Category_2"].fillna(df["Product_Category_2"].mode()[0])
```

```python
df["Product_Category_2"].isnull().sum()
```
