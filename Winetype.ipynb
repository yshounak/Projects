### 11.Find correlation between citric acid content and PH.

### 12.Find regression of PH on citric acid

```python
from sklearn.linear_model import LinearRegression

# Extract the features (citric acid) and target (pH)
X = combined_df[['citric acid']]
y = combined_df['pH']

# Initialize and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Extract the coefficients
slope = model.coef_[0]
intercept = model.intercept_

# Print the regression equation
print(f"Regression equation: pH = {slope:.2f} * citric acid + {intercept:.2f}")

# Visualize the regression line
plt.scatter(X, y, alpha=0.5)
plt.plot(X, model.predict(X), color='red')
plt.xlabel('Citric Acid')
plt.ylabel('pH')
plt.title('Regression of pH on Citric Acid')
plt.show()
```

### 13.Find how many variables are independent.

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Set a threshold for correlation values to determine independence
threshold = 0.5  # Adjust as needed

# Count the number of independent variables
num_independent_variables = 0
for column in correlation_matrix.columns:
    if (correlation_matrix[column].abs() < threshold).all():
        num_independent_variables += 1

print("Number of independent variables:", num_independent_variables)
```

### 22.Is the confidence interval for the differences of the proportion of white wine with rating 3 and the proportion of red wine with rating 3.



```python
from scipy.stats import norm

# Calculate the proportions of white and red wines with rating 3
white_count_3 = combined_df[(combined_df['color'] == 'white') & (combined_df['quality'] == 3)].shape[0]
red_count_3 = combined_df[(combined_df['color'] == 'red') & (combined_df['quality'] == 3)].shape[0]

total_white = combined_df[combined_df['color'] == 'white'].shape[0]
total_red = combined_df[combined_df['color'] == 'red'].shape[0]

prop_white_3 = white_count_3 / total_white
prop_red_3 = red_count_3 / total_red

# Calculate the standard error
se = np.sqrt((prop_white_3 * (1 - prop_white_3) / total_white) + (prop_red_3 * (1 - prop_red_3) / total_red))

# Set the confidence level (e.g., 95%)
confidence_level = 0.95

# Calculate the z-score for the confidence level
z_score = norm.ppf((1 + confidence_level) / 2)

# Calculate the margin of error
margin_of_error = z_score * se

# Calculate the confidence interval
difference = prop_white_3 - prop_red_3
lower_bound = difference - margin_of_error
upper_bound = difference + margin_of_error

print("Confidence Interval for the Difference of Proportions:")
print(f"({lower_bound}, {upper_bound})")
```
